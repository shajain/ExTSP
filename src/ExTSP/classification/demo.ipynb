```python
from ExTSP.classification.auc_bootstrap import bootstrap_auc_for_disease
from ExTSP.config import Diseases
for disease in ["PKD", "PH", "PCD", "ASD"]:
    print(f"================{disease}=================")
    for i in range(1,6):
        if disease == "ASD" and i > 3:
            continue
        print(f"Setting {i}")
        result1, summary1 = bootstrap_auc_for_disease(disease, i, n_bootstrap=100, prop_gene_draws=1.0)
        print(f"\tStandard: IsoPath: {summary1['mean_auc_isopath']:.3f}, exTSP: {summary1['mean_auc_extsp']:.3f}")
        if i != 3:
            result2, summary2 = bootstrap_auc_for_disease(disease, i, balanced_negative=True, n_bootstrap=100, prop_gene_draws=1.0)
            print(f"\tBalanced: IsoPath: {summary2['mean_auc_isopath']:.3f}, exTSP: {summary2['mean_auc_extsp']:.3f}")
```
